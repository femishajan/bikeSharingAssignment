```python
#creating dummy variables

months=pd.get_dummies(boombike.month,drop_first=True)
weekdays=pd.get_dummies(boombike.weekday,drop_first=True)
weather_sit=pd.get_dummies(boombike.weathersit,drop_first=True)
seasons=pd.get_dummies(boombike.season,drop_first=True)
```

```python
#Function for checking VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif():
    vif = pd.DataFrame()
    X = X_train_rfe
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif
```

```python
#checking VIF
calculate_vif()
```
